```python
from re import VERBOSE
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.initializers import RandomNormal,HeNormal
from tensorflow.keras import Input
import numpy as np
import matplotlib.pyplot as plt
import random
import keras

data_in = []
data_out = []

with open("/content/base_datos_Eva1.csv", "r") as archivo:
    filas = archivo.readlines()
    filas.pop(0)
    random.seed(42)
    random.shuffle(filas)

    for linea in filas:
        linea = linea.rstrip('\n')
        valores = linea.split(',')
        temp_int, temp_ext, hum, es_dia, ventana, ventilador = valores

        valor_es_dia = 1.0 if es_dia.strip().lower() == "true" else 0.0
        data_in.append([float(temp_int), float(temp_ext), float(hum), valor_es_dia])

        valor_ventana = 1.0 if ventana.strip().lower() == "true" else 0.0
        valor_ventilador = 1.0 if ventilador.strip().lower() == "true" else 0.0
        data_out.append([valor_ventana, valor_ventilador])

regla_muestras = int(len(data_in) * 0.7)

in_entrenamiento = data_in[0:regla_muestras]
in_validacion = data_in[regla_muestras:]
out_entrenamiento = data_out[0:regla_muestras]
out_validacion = data_out[regla_muestras:]


modelo = Sequential()
modelo.add(Input(shape=(4,)))
modelo.add(Dense(units=2, activation="sigmoid"))



modelo.compile(loss='binary_crossentropy',
               optimizer=Adam(learning_rate=0.01),
               metrics = ['accuracy'])

historico = modelo.fit(x=np.array(in_entrenamiento),
                       y=np.array(out_entrenamiento),
                       epochs=100,
                       verbose=1,
                       validation_data=(np.array(in_validacion), np.array(out_validacion)))

modelo.save("modelo_reles.h5")

```
